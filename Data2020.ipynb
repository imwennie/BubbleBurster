```python
# Load the senate list file which contain senate information, 
# Name, Party, State and Twitter Account for all three election year 
data = pd.read_csv('Updated_Senate_List.csv', sep = ',')
data.head(5)
```

```python
# we want to focus on 2020
data[['Current Senate', 'Party.2', 'State.2', 'Twitter Account.2']]
```

# 2020 Election Year

```python
# 2020 Election year
account2020 = list(data['Twitter Account.2'].values)
tweets2020 = pd.DataFrame(columns = ['Datetime', 'Text','TweetID','username','geo'])

for account in account2020:
    print(account)
    if account == 'NONE':
        print('\t Skip scraping because of invalid twitter account input')
        continue
    else:
        tweetCriteria = got.manager.TweetCriteria().setUsername(account)\
                                           .setSince('2020-01-01')\
                                           .setUntil('2020-05-29')\
                                           .setMaxTweets(5000)
        tweets = got.manager.TweetManager.getTweets(tweetCriteria)
        tweets_detail = [[tweet.date, tweet.text,tweet.id,tweet.username,tweet.geo ] for tweet in tweets]
        tweets2020 = tweets2020.append(pd.DataFrame(tweets_detail, 
                                                    columns = ['Datetime', 'Text','TweetID','username','geo']), 
                                       ignore_index = True)
        print('\t Done scraping')
```
