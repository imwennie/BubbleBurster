# 2012 Election Year

```python
# 2012 Election year
electionday2012 = '2012-11-06'
account2012 = list(data['Twitter Account'].values)
tweets2012 = pd.DataFrame(columns = ['Datetime', 'Text','TweetID','username','geo'])

for account in account2012:
    print(account)
    if account == 'NONE':
        print('\t Skip scraping because of invalid twitter account input')
        continue
    else:
        tweetCriteria = got.manager.TweetCriteria().setUsername(account)\
                                           .setSince('2012-07-01')\
                                           .setUntil(electionday2012)\
                                           .setMaxTweets(5000)
        tweets = got.manager.TweetManager.getTweets(tweetCriteria)
        tweets_detail = [[tweet.date, tweet.text,tweet.id,tweet.username,tweet.geo ] for tweet in tweets]
        tweets2012 = tweets2012.append(pd.DataFrame(tweets_detail, 
                                                    columns = ['Datetime', 'Text','TweetID','username','geo']), 
                                       ignore_index = True)
        print('\t Done scraping')
```
